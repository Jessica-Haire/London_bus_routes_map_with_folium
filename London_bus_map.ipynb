```python
import geopandas as gpd
import folium
from folium import GeoJson, Tooltip

buses_gdf = gpd.read_file("busroutes.shp")

# Convert GeoDataFrame to GeoJSON
geojson_data = buses_gdf.__geo_interface__

# Create a Folium map centered on London
m = folium.Map(location=[51.5074, -0.1278], zoom_start=10)

# Add bus routes as line overlays with tooltips
for feature in geojson_data['features']:
    geometry = feature['geometry']
    properties = feature['properties']
    route_name = properties['Route']
    length_km = properties['LengthKm']
    coordinates = geometry['coordinates']
    line = GeoJson(geometry, tooltip=f"Route: {route_name}, Length: {length_km} km")
    line.add_to(m)

# Add a layer control to toggle the visibility of different overlays
folium.LayerControl().add_to(m)

# Display the map
m
```
